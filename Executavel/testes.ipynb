```python
import csv

# Função para classificar um exame
def classificar_exame(exame_id):
    print(f"Exame ID: {exame_id}")
    classes = {
        "ST": "ST",
        "SB": "SB",
        "AF": "AF",
        "LBBB": "LBBB",
        "RBBB": "RBBB",
        "1DAVB": "1dAVb",
        "NORMAL": "normal"
    }
    
    print("Classes disponíveis:")
    for classe in classes.values():
        print(classe)
    
    classificacao = ""
    while classificacao not in classes:
        classificacao = input(f"Escolha uma classe para o exame {exame_id}: ").strip().upper()
        if classificacao not in classes:
            print("Classe inválida. Tente novamente.")
    
    return classes[classificacao]

# Lista de IDs de exames (exemplo)
exames_ids = ["exame1", "exame2", "exame3"]

# Dicionário para armazenar as classificações
resultados = []

# Classificar cada exame
for exame_id in exames_ids:
    classificacao = classificar_exame(exame_id)
    resultado = {
        "ID": exame_id,
        "ST": classificacao == "ST",
        "SB": classificacao == "SB",
        "AF": classificacao == "AF",
        "LBBB": classificacao == "LBBB",
        "RBBB": classificacao == "RBBB",
        "1dAVb": classificacao == "1dAVb",
        "normal": classificacao == "normal"
    }
    resultados.append(resultado)

# Nome do arquivo CSV
csv_file = "classificacoes_exames.csv"

# Escrever os resultados no arquivo CSV
with open(csv_file, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["ID", "ST", "SB", "AF", "LBBB", "RBBB", "1dAVb", "normal"])
    writer.writeheader()
    writer.writerows(resultados)

print(f"As classificações foram salvas no arquivo {csv_file}")
```

```python
import csv
import tkinter as tk
from tkinter import simpledialog
from PIL import Image, ImageTk

# Função para classificar um exame
def classificar_exame(exame_id, img_path):
    # Cria a janela principal
    root = tk.Tk()
    root.withdraw()  # Esconde a janela principal

    # Mostra a imagem do exame
    img = Image.open(img_path)
    img.show()

    # Dicionário de classes
    classes = {
        "ST": "ST",
        "SB": "SB",
        "AF": "AF",
        "LBBB": "LBBB",
        "RBBB": "RBBB",
        "1DAVB": "1dAVb",
        "NORMAL": "normal"
    }

    # Mostra um diálogo para o usuário escolher a classe
    classificacao = ""
    while classificacao not in classes:
        classificacao = simpledialog.askstring(
            "Classificação de Exame",
            f"Escolha uma classe para o exame {exame_id} (ST, SB, AF, LBBB, RBBB, 1dAVb, normal):"
        ).strip().upper()
        if classificacao not in classes:
            print("Classe inválida. Tente novamente.")

    return classes[classificacao]

# Lista de IDs de exames e caminhos para as imagens (exemplo)
exames = [
    {"id": "exame1", "img_path": "ecg(1748223).png"},
    {"id": "exame2", "img_path": "ecg(1748223).png"},
    {"id": "exame3", "img_path": "ecg(1748223).png"}
]

# Dicionário para armazenar as classificações
resultados = []

# Classificar cada exame
for exame in exames:
    classificacao = classificar_exame(exame["id"], exame["img_path"])
    resultado = {
        "ID": exame["id"],
        "ST": classificacao == "ST",
        "SB": classificacao == "SB",
        "AF": classificacao == "AF",
        "LBBB": classificacao == "LBBB",
        "RBBB": classificacao == "RBBB",
        "1dAVb": classificacao == "1dAVb",
        "normal": classificacao == "normal"
    }
    resultados.append(resultado)

# Nome do arquivo CSV
csv_file = "classificacoes_exames.csv"

# Escrever os resultados no arquivo CSV
with open(csv_file, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["ID", "ST", "SB", "AF", "LBBB", "RBBB", "1dAVb", "normal"])
    writer.writeheader()
    writer.writerows(resultados)

print(f"As classificações foram salvas no arquivo {csv_file}")
```
