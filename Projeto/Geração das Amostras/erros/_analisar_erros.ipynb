```python
import pandas as pd

# Função para adicionar e preencher as colunas
def add_disease_columns(df, disease):
    df['1dAVb'] = False
    df['RBBB'] = False
    df['LBBB'] = False
    df['SB'] = False
    df['ST'] = False
    df['AF'] = False
    df[disease] = True  # Define a coluna da doença atual como True
    return df

# Carregar os DataFrames
SB_merged = pd.read_csv('SB.csv')
ST_merged = pd.read_csv('ST.csv')
AF_merged = pd.read_csv('AF.csv')
dAVB_merged = pd.read_csv('1dAVB.csv')
RBBB_merged = pd.read_csv('RBBB.csv')
LBBB_merged = pd.read_csv('LBBB.csv')

# Adicionar e preencher as colunas de doença
SB_merged = add_disease_columns(SB_merged, 'SB')
ST_merged = add_disease_columns(ST_merged, 'ST')
AF_merged = add_disease_columns(AF_merged, 'AF')
dAVB_merged = add_disease_columns(dAVB_merged, '1dAVb')
RBBB_merged = add_disease_columns(RBBB_merged, 'RBBB')
LBBB_merged = add_disease_columns(LBBB_merged, 'LBBB')

# Filtrar os erros para cada DataFrame
SB_erros = SB_merged[(SB_merged['normal_ecg'] == 0) & (SB_merged['doenca'] != 'SB')]
ST_erros = ST_merged[(ST_merged['normal_ecg'] == 0) & (ST_merged['doenca'] != 'ST')]
AF_erros = AF_merged[(AF_merged['normal_ecg'] == 0) & (AF_merged['doenca'] != 'AF')]
dAVB_erros = dAVB_merged[(dAVB_merged['normal_ecg'] == 0) & (dAVB_merged['doenca'] != '1dAVb')]
RBBB_erros = RBBB_merged[(RBBB_merged['normal_ecg'] == 0) & (RBBB_merged['doenca'] != 'RBBB')]
LBBB_erros = LBBB_merged[(LBBB_merged['normal_ecg'] == 0) & (LBBB_merged['doenca'] != 'LBBB')]

# Combinar todos os DataFrames de erros
merged_df  = pd.concat([SB_erros, ST_erros, AF_erros, dAVB_erros, RBBB_erros, LBBB_erros])
# Reorder the columns
merged_df = merged_df[['exam_id', 'label', '1dAVb', 'RBBB', 'LBBB', 'SB', 'ST', 'AF', 'normal_ecg', 
                       'silhouette_score', 'tipo', 'doenca', 'borderline', 'comentario']]

# Save to CSV
merged_df.to_csv('erros.csv', index=False)
print("Arquivo 'erros.csv' salvo com sucesso.")
```
