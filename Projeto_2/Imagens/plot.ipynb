```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.patches as mpatches

# Carregar as imagens
images = ['ST.png', 'SB.png', 'AF.png', 'RBBB.png', 'LBBB.png']
loaded_images = [mpimg.imread(image) for image in images]

# Configurar a figura e os subplots
fig, axes = plt.subplots(1, 5, figsize=(20, 4))  # 5 imagens lado a lado

# Exibir cada imagem em um subplot
for ax, img, title in zip(axes, loaded_images, images):
    ax.imshow(img)
    ax.axis('off')  # Esconder os eixos
    ax.set_title(title.split('.')[0])  # Definir o título de cada imagem sem extensão

# Criar a legenda universal com fonte maior
legend_patches = [
    mpatches.Patch(color='green', label='Disease-pattern Cluster'),
    mpatches.Patch(color='navy', label='Normal Cluster')
]

# Adicionar a legenda centralizada na parte inferior da figura
fig.legend(handles=legend_patches, loc='center', bbox_to_anchor=(0.5, -0.05), ncol=2, fontsize='large')

# Ajustar o espaçamento entre os subplots
plt.subplots_adjust(wspace=0.05, bottom=0.2)  # Reduzir o espaçamento horizontal e ajustar o espaçamento inferior

# Salvar a figura concatenada, ajustando para remover margens desnecessárias
plt.savefig('concatenated_images.png', bbox_inches='tight', pad_inches=0.1)
plt.show()
```
